# Interthread Communication by using Condition object

```python
from threading import *

def consumer():
    c.acquire()
    print('Consumer is waiting items.')
    c.wait(10)
    print('Consumer got notifications and  consuming items.')
    c.release()
    
def producer():
    c.acquire()
    print('Producer is producing items.')
    print('Producer is giving notifications.')
    c.notify()
    c.release()

c=Condition()

t1=Thread(target=consumer)
t2=Thread(target=producer)

t1.start()
t2.start()
```

```python
from threading import *
import time
import random

items=[]

def producer():
    while True:
        c.acquire()
        item=random.randint(1,100)
        print('Producer is producing items:',item) #2
        items.append(item)
        print('Producer is giving notifications.') #3
        c.notify()
        c.release()
        time.sleep(3)

def consumer():
    while True:
        c.acquire()
        print('Consumer is waiting for updation.') #1
        c.wait(10)
        while len(items) !=0:
            print('Consumer got notifications and  consuming items:',items.pop()) #4
        c.release()
        time.sleep(3)

c=Condition()

t1=Thread(target=consumer)
t2=Thread(target=producer)

t1.start()
t2.start()
```

# Interthread Communication by using Queue object

```python
from threading import *
import time
import random
import queue

def produce(q):
    while True:
        item=random.randint(1,100)
        print('Producer is producing items:',item) #2
        q.put(item)
        print('Producer is giving notifications.') #3
        time.sleep(3)

def consume(q):
    while True:
        print('Consumer is waiting for updation.') #1
        print('Consumer got notifications and  consuming items:',q.get()) #4
        time.sleep(3)

q=queue.Queue()

t1=Thread(target=consume,args=(q,))
t2=Thread(target=produce,args=(q,))

t1.start()
t2.start()
```

# Types of Queue
1. FIFO
2. LIFO
3. Priority

# FIFO Queue

```python
import queue
q=queue.Queue()
q.put(10)
q.put(5)
q.put(20)
q.put(15)
while not q.empty():
    print(q.get(),end=' ')
```

# LIFO Queue

```python
import queue
q=queue.LifoQueue()
q.put(10)
q.put(5)
q.put(20)
q.put(15)
while not q.empty():
    print(q.get(),end=' ')
```

# Priority Queue

```python
import queue
q=queue.PriorityQueue()
#q.put(priority,element)
q.put((1,'25'))
q.put((4,'20'))
q.put((2,'15'))
q.put((3,'15'))
q.put((8,'A'))
q.put((7,'B'))
q.put((6,'C'))
q.put((5,'C'))

while not q.empty():
    print(q.get(),end=' ')
  
```

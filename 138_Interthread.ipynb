# Interthread Communication by using Event Object

```python
from threading import *
import time

def producer():
    time.sleep(10)
    print('Producer thread producing items') #2
    print('Producer thread giving notification') #3
    e.set()

def consumer():
    print('consumer thread is waiting for updation') #1
    e.wait()
    print('consumer thread consuming items') #4

t1=Thread(target=producer)
t2=Thread(target=consumer)

e=Event()
t1.start()
t2.start()
```

```python
from threading import *
import time

def trafficpolice():
    while True:
        time.sleep(10)
        print('Traffic Police giving GREEN signal') #2
        e.set()
        time.sleep(20)
        print('Traffic Police giving RED signal') #5
        e.clear()
        
def vehicles():
    num=0
    while True:
        print('Vehicles are waiting for GREEN signal') #1
        e.wait()
        print('Traffic Signal is GREEN, Vehicles can move') #3
        while e.is_set():
            num=num+1
            print(f'Vehicle no:{num}, Crossing the signal') #4
            time.sleep(2)

        print('Traffic Signal is RED...Vehicles has to wait.') #6 #2

e=Event()
t1=Thread(target=trafficpolice)
t2=Thread(target=vehicles)
t1.start()
t2.start()
```
